```python
# 1. 넘파이(NumPy)와 맷플롯립(Matplotlib) 불러오기
# 넘파이는 np, 맷플롯립은 plt라는 짧은 별명으로 부르겠습니다.
import numpy as np
import matplotlib.pyplot as plt

# 2. 중력 가속도 설정
g = 9.8

# 3. 시간 데이터 만들기 (NumPy 활용)
# np.linspace(시작, 끝, 개수)는 시작부터 끝까지 지정한 개수만큼 잘게 쪼개줍니다.
# 0초부터 10초까지 100개의 구간으로 쪼개서 t라는 변수에 저장해 보세요.
t = np.linspace(0, 10, 100)

print("만들어진 시간 데이터의 개수:", len(t))
```

```python
# 초기 속도와 발사 각도 설정
v0 = 50      # 대포알의 초기 속도 (m/s)
angle = 45   # 발사 각도 (도)

# 파이썬에서 삼각함수를 계산할 때는 우리가 아는 '도(degree)' 대신 '라디안(radian)'을 씁니다.
# 각도를 라디안으로 변환합니다.
theta = np.radians(angle)

# --- x축 방향 계산 (가로 방향으로는 일정한 속도로 날아갑니다: 등속 운동) ---
# 가로 방향 초기 속도 = v0 * cos(theta)
v_x = v0 * np.cos(theta)

# 가로 방향 위치 = 속도 * 시간
x = v_x * t

# --- y축 방향 계산 (세로 방향으로는 중력을 받아 속도가 줄어듭니다: 등가속도 운동) ---
# 세로 방향 초기 속도 = v0 * sin(theta)
v_y = v0 * np.sin(theta)

# 세로 방향 위치 = (초기 속도 * 시간) - (0.5 * 중력가속도 * 시간의 제곱)
y = v_y * t - 0.5 * g * (t ** 2)

# --- 땅(y=0) 밑으로 떨어진 데이터는 잘라내기 ---
x = x[y >= 0]
y = y[y >= 0]

# 그래프 그리기
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='green', linewidth=2)
plt.title('Projectile Motion')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.grid(True) # 격자 무늬 표시
plt.show()
```

```python
plt.figure(figsize=(8, 4))

# 대포를 쏠 세 가지 각도를 리스트로 만듭니다.
angles = [30, 45, 60]

# for문을 이용해 각도를 하나씩 꺼내면서 그래프를 겹쳐서 그립니다.
for a in angles:
    # 1. 각도 변환
    theta = np.radians(a)

    # 2. 가로, 세로 속도 계산
    v_x = v0 * np.cos(theta)
    v_y = v0 * np.sin(theta)

    # 3. 가로, 세로 위치 계산
    x = v_x * t
    y = v_y * t - 0.5 * g * (t ** 2)

    # 4. 땅 위(y >= 0)의 데이터만 남기기
    x = x[y >= 0]
    y = y[y >= 0]

    # 그래프에 선을 그리고, 어떤 각도인지 라벨(label)을 붙입니다.
    plt.plot(x, y, label=f'{a} degrees')

plt.title('Which angle goes furthest?')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.legend() # 화면에 라벨(범례)을 표시합니다.
plt.grid(True)
plt.show()
```
